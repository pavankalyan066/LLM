```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# download and load the model
torch.set_default_device('cuda')
model = AutoModelForCausalLM.from_pretrained("microsoft/phi-1_5", trust_remote_code=True, torch_dtype="auto")
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-1_5", trust_remote_code=True, torch_dtype="auto")
```

```python
inputs = tokenizer('''what are the top electric vechile companies. Answer:''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=100)
text = tokenizer.batch_decode(outputs)[0]
print(text)
```

```python
inputs = tokenizer('''what is the capital of india. Answer:''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=100, temperature=0)
text = tokenizer.batch_decode(outputs)[0]
print(text)
```

```python
inputs = tokenizer('''write a joke about elon musk. Answer:''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=100)
text = tokenizer.batch_decode(outputs)[0]
print(text)
```

```python
text = """The Company believes it has a talented, motivated, and dedicated team, and is committed to supporting the development of all
of its team members and to continuously building on its strong culture. As of September  25, 2021, the Company had
approximately 154,000 full-time equivalent employees.
from the given text what is count of the employees. Answer:
"""
```

```python
inputs = tokenizer(text, return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=100)
text = tokenizer.batch_decode(outputs)[0]
print(text)
```
